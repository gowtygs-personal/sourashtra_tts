```python
import torch
import IPython
from TTS.api import TTS
from TTS.tts.configs.xtts_config import XttsConfig
```

```python
# Get device
#device = "cuda" if torch.cuda.is_available() else "cpu"
device = "cuda"
# Required to enable loading model on Pytorch 2.6 and Above
torch.serialization.add_safe_globals([XttsConfig])
# Init TTS
```

```python
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2")
tts.to(device=device)
```

```python
# Run TTS
# â— Since this model is multi-lingual voice cloning model, we must set the target speaker_wav and language
# Text to speech list of amplitude values as output
wav = tts.tts(text="This is a test. I'm trying to test the inference speed with GPU", speaker_wav="models/en_sample.wav", language="en")
```

```python
# Text to speech to a file
tts.tts_to_file(text="This is a test. I'm trying to test the inference speed with GPU", speaker_wav="models/en_sample.wav", language="en", file_path="output/trial_1.wav")
```
